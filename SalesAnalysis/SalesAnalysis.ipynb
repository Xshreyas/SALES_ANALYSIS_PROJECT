# Importing Required Libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

#### Merge the 12 months of sales data into a single CSV file

```python
path = "./Sales_Data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data_copy.csv", index=False)
    
```

#### Read the updated dataframe

```python
all_data = pd.read_csv("all_data_copy.csv")
all_data.head()
```

## Clean-up the data !!

### Drop rows of NAN

```python
# Find NAN
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data.head()
```

##### Get rid of text in order date column

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or'
```

#### Make columns correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Augment data with additional columns !!


#### Add month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### Add City column

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
all_data.head()
```

### Question 1 : What was the best month for sales? How much was earned that month ?

```python
all_data['Sales'] = all_data['Quantity Ordered'].astype('int') * all_data['Price Each'].astype('float')
```

```python
all_data.groupby(['Month']).sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)
print(months)

plt.bar(months,all_data.groupby(['Month']).sum()['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

### Question 2 : Which city sold the most product ?

```python
results = all_data.groupby(['City']).sum()
results
```

```python
import matplotlib.pyplot as plt

all_data['City'] = all_data['City'].astype(str)
city_sales = all_data.groupby('City')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(city_sales.index, city_sales.values)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.xticks(rotation='vertical', size=8)
plt.title('Total Sales by City')
plt.show()
```

### Question 3 : What time should we display advertisements to maximize likelihood of customer's buying product ?

#### Add Hour column

```python

all_data['Hour'] = pd.to_datetime(all_data['Order Date']).dt.hour
all_data['Minute'] = pd.to_datetime(all_data['Order Date']).dt.minute
all_data['Count'] = 1
all_data.head()
```

```python
hours = all_data.groupby('Hour').size()

plt.plot(hours.index, hours.values)
plt.xticks(hours.index)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.title('Orders by Hour')
plt.grid()
plt.show()
```

### Question 4 : What products are most often sold together ?

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)].copy()

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    
    items = row.split(',')
    
    count.update(Counter(combinations(items, 2)))

for (prod1, prod2), freq in count.most_common(10):
    print(f"{prod1.strip()} & {prod2.strip()}  sold together {freq} times")

```
